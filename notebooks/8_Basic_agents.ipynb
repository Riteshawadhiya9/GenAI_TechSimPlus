```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_core.tools import tool
```

```python
@tool
def add_num(a:int, b : int) :
    """
    it will return the sum of a and b
    Args:
    a (int): first number
    b (int): second number 

    """
    return a + b


@tool
def multiply_num(a:int, b : int) :
    """
    it will return the product of a and b
    Args:
    a (int): first number
    b (int): second number 

    """
    return a * b
```

```python
add_num.invoke({"a": 5, "b": 10})
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain.agents import create_agent
```

```python
llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash" )
```

```python
agent = create_agent(
    model=llm,
    tools=[add_num , multiply_num],
    system_prompt= " you are a maths teacher, and always use tool for calculation  "
)
```

```python
response = agent.invoke({"messages": [{"role": "user", "content": "what is 5*2+10 ?"}]})
```
