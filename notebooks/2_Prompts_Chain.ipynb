## Prompt Template - Chain of Prompts (Dynamic Prompts)

```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
```

```python
llm  = ChatGoogleGenerativeAI(
    model ="gemini-2.5-flash"
)
out = StrOutputParser()
```

```python
prompt = ChatPromptTemplate.from_messages([
    {"role" : "system" , "content":"You are a Translater and translate in {language}"},
    {"role" : "user" , "content":"{query}"}
]) 
prompt   
```

## My Self Chain Creation (Node)

```python
def transform_case(result:str):
    return result.upper()
```

```python
chains = prompt | llm | out | transform_case
```

```python
res = chains.invoke({"language" : "Hinglish" , "query" : "I Love you kratika❤💋"})
# print(res.content) // If you use StrOutputParser (out) then no need to use .content
print(res)
```
