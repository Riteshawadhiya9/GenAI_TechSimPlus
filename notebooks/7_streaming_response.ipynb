## Use Streaming Response in Gemini Model 

```python
from langchain_google_genai import ChatGoogleGenerativeAI
llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash" , streaming=True)
```

```python
ques = "what is genAI?"
```

```python
res = llm.stream(ques)
```

```python
for chunk in res: # if you want to run again then you have to run the above code again because the streaming memory is cleared after the response is generated 
    # print(chunk, "\n")
    print(chunk.content, end="")
```

## Use Streaming Response in Groq Model 

```python
from langchain_groq import ChatGroq
llm = ChatGroq(model="openai/gpt-oss-20b" , streaming=True)
```

```python
ques = "can you explain me genAI in detail? "
res = llm.stream(ques)
```

```python
for chunk in res:
    print(chunk.content, end="")
```
