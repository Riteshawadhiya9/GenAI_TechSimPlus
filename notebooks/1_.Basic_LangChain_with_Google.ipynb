##### Connect to Google Gemini Model via LangChain

```python
from langchain_google_genai import ChatGoogleGenerativeAI

llm = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash",
    temperature=0.7
)
```

```python
response = llm.invoke("Write a Hello World program in Java and Python.")
# response
# response.content
# print(response)
print(f"Answer :-  {response.content}")
```

# Static Prompts

```python
from langchain_google_genai import ChatGoogleGenerativeAI
llm = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash",
    temperature=0.7
)
```

```python
prompt = [
    ("system" , "you are a java developer.") ,## "system -> act as a ..." , "ai" , "user"
    ("user" , "what is array with example & syntax?") 
]
response = llm.invoke(prompt)
print(f"Answer -> {response.content}")
```
