```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash")
```

>### Discuss the Problem First

```python
text = "My name is Ritesh Awadhiya , Email is riteshawadhiya723@gmail.com and my age is 21"
```

```python
res = llm.invoke(f"give only name,email,age from this text : {text}")
print(res.content)
```

 
>### Solution is : Structure Output

```python
from pydantic import BaseModel , Field
from typing import List

class ResponseStructure(BaseModel):
    name: str = Field( description="Name of the person")
    email: str = Field( description="Email of the person")
    age: int = Field( description="Age of the person")

structure_llm = llm.with_structured_output(ResponseStructure)
    
```

```python
res = structure_llm.invoke(f"give only name,email,age from this text : {text}")
# res -> ResponseStructure(name='Ritesh Awadhiya', email='riteshawadhiya723@gmail.com', age=21)

print(res.name) # Ritesh Awadhiya
print(res.email) # riteshawadhiya723@gmail.com
print(res.age) # 21

res.model_dump() # this converts pydantic class obj into dictionary -> {'name': 'Ritesh Awadhiya', 'email': 'riteshawadhiya723@gmail.com', 'age': 21}
```

```python
class Movies(BaseModel):
    name: str = Field( description="Name of the Movie")
    director: str = Field( description="Director of the Movie")
    release_year: int = Field( description="Release Year of the Movie")

class MoviesList(BaseModel):
    movies: List[Movies] = Field( description="List of Movies")
movie_llm = llm.with_structured_output(MoviesList)    
```

```python
res = movie_llm.invoke("give me three hollywood movies")
```
